```python
script = '''import streamlit as st
from simpletransformers.classification import ClassificationModel
import numpy as np
import translators as ts
import translators.server as tss


decode_dict =  {0: 'action',
                             1: 'adventure',
                             2: 'animation',
                             3: 'biography',
                             4: 'comedy',
                             5: 'crime',
                             6: 'documentary',
                             7: 'drama',
                             8: 'family',
                             9: 'fantasy',
                             10: 'horror',
                             11: 'musical',
                             12: 'mystery',
                             13: 'noir',
                             14: 'romance',
                             15: 'science fiction',
                             16: 'suspense',
                             17: 'thriller',
                             18: 'war',
                             19: 'western'}
                             
                             
class genre_predictor:

    def __init__(self):
        self.model = ClassificationModel('bert', '/content/drive/MyDrive/Colab/outputs', use_cuda = True)
        
        self.first_genre = None
        self.second_genre = None
        
    def predict(self, your_text):
        predictions, raw_outputs = self.model.predict([your_text])
        self.first_genre = decode_dict[predictions[0]]
        
        raw_outputs[0][predictions[0]] = 0
        self.second_genre = decode_dict[np.argmax(raw_outputs)]
        
        return self.first_genre, self.second_genre
        
predictor = genre_predictor()



lang = st.selectbox(
        "–í—ã–±–µ—Ä–∏—Ç–µ —è–∑—ã–∫",
        ("English", "–†—É—Å—Å–∫–∏–π"),
    )




x = st.text_area('–û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å –∂–∞–Ω—Ä–∞ 9000')  # üëà this is a widget
if st.button('–ù–∞–∂–º–∏ –Ω–∞ –∫–Ω–æ–ø–∫—É - –ø–æ–ª—É—á–∏—à—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç'):
  if len(x) == 0:
      st.write('–í–≤–µ–¥–∏—Ç–µ —Ç–µ–∫—Å—Ç –∏ –Ω–∞–∂–º–∏—Ç–µ –∫–Ω–æ–ø–∫—É')
      
  else:
      if lang == '–†—É—Å—Å–∫–∏–π':
          x = tss.translate_text(x, 'google')
          st.write('–ß—Ç–æ–∂, –º–µ—Å—å–µ, —è –ø–µ—Ä–µ–≤–µ–¥—É –Ω–∞ —è–∑—ã–∫ –º–æ–¥–µ–ª–∏ –∑–∞ –í–∞—Å')
      st.write('–í–∞—à —Ç–µ–∫—Å—Ç', x)
      st.write('–ù–µ–º–Ω–æ–≥–æ –ø–æ–¥–æ–∂–¥–µ–º...')
      
      first_genre, second_genre = predictor.predict(x)
      st.write('–ù–∞–∏–±–æ–ª–µ–µ –≤–µ—Ä–æ—è—Ç–Ω—ã–π –∂–∞–Ω—Ä:', first_genre)
      st.write('–ù–æ —Ç–∞–∫–∂–µ –≤–æ–∑–º–æ–∂–Ω–æ:', second_genre)
'''
with open('test.py', 'w') as f:
    f.write(script)
```
